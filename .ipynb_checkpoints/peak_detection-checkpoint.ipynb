```python
import json
import numpy as np
import matplotlib.pyplot as plt
import sounddevice as sd

from scipy import signal
```

```python
with open('sample.txt', 'r') as f:
    data = json.load(f)
audio_full = np.array(data['audio'])
fs = data['fs']
audio_full.shape
```

```python
plt.plot(audio_full)
```

```python
audio = audio_full[1000000:1000000 + 10*fs]
l = len(audio)
plt.plot(audio)
```

```python
fft_n = 2**int(np.ceil(np.log2(l)))
fft_n, l
```

```python
fft_audio = np.fft.fft(audio, n=fft_n)[fft_n//2:]
f_axis = np.linspace(0, fs//2, fft_n//2)
plt.plot(f_axis[:300], abs(fft_audio[:300]))
```

```python
sd.play(audio)
```

```python
sd.stop()
```
