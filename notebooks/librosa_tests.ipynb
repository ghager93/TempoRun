```python
import os
import numpy as np
import librosa
import pydub
import matplotlib.pyplot as plt
```

```python
AUDIOPATH = "/home/ghager/dev/learning/TempoRun/audio/06 We Could Forever.m4a"
```

```python
audio = pydub.AudioSegment.from_file(AUDIOPATH)
y, sr = np.array(audio.get_array_of_samples()).astype(np.float32), audio.frame_rate
```

```python
tempo, beats = librosa.beat.beat_track(y=y, sr=sr)
```

```python
onset_env = librosa.onset.onset_strength(y=y, sr=sr)
pulse = librosa.beat.plp(onset_envelope=onset_env, sr=sr)
beats_plp = np.flatnonzero(librosa.util.localmax(pulse))
times = librosa.times_like(onset_env, sr=sr)
```

```python
plt.plot(times, librosa.util.normalize(pulse))
plt.vlines(times[beats_plp], 0, 1, alpha=0.5, color='r', linestyles='--', label='Beats')
plt.xlim([100, 110])
```
